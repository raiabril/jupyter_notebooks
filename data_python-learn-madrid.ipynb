## EXTRAER

```python
file = open('FMEL_Dataset.csv', mode = 'r', encoding= 'UTF-8')
```

```python
data = pd.read_csv(file, delimiter = ',')
```

```python
file.close()
```

## TRANSFORM DATA

```python
data.set_index('id')
data['year'] = data.loc[:,'season'].str.split('-').str[0]
```

### Goles por temporada

```python
goals_per_season = data.groupby('year').sum()
```

```python
goals_per_season.head()
```

```python
plt.plot(goals_per_season.index,goals_per_season.localGoals)
plt.xlabel('localGoals')
plt.ylabel('visitorGoals')
plt.title('Goles por temporada')
plt.show()
```

## Plot local vs visitor goals

```python
plt.scatter(goals_per_season.localGoals, goals_per_season.visitorGoals)
plt.xlim([200,1500])
plt.ylim([200,1500])
plt.xlabel('localGoals')
plt.ylabel('visitorGoals')
plt.title('Goles local vs visitante')
plt.show()
```
